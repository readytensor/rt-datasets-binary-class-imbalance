```python
import numpy as np
import pandas as pd
import os
```

```python
dataset_name = 'credit_approval'
```

```python
input_dir = './raw'
inp_fname = "crx.data"
output_dir = f'./../../processed/{dataset_name}/'
outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
```

# Read Data

```python
columns = [ 
    "A1",
    "A2",
    "A3",
    "A4",
    "A5",
    "A6",
    "A7",
    "A8",
    "A9",
    "A10",
    "A11",
    "A12",
    "A13",
    "A14",
    "A15",
    "A16",
]
columns
```

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname), names=columns) 
print(data.shape)
data.head()
```

```python
id_col = "id"
target_col = "A16"
```

# Insert Id Column

```python
# insert Id column 
if id_col not in data.columns:
    N = data.shape[0]
    data.insert(0, id_col, np.arange(N))
    print(data.head())
```

# Replace null indicators

```python
data = data.replace('?', np.nan)
```

# Update Class labels

```python
mapping = {
    "+": "positive", 
    "-": "negative"
}

data['A16'] = data['A16'].map(mapping)
data.head()
```

# Save Main Data File

```python
data.to_csv(outp_fname, index=False)
```
